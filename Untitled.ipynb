```python
import matplotlib.pyplot as plt
import numpy as np
import numpy.random
from scipy.fftpack import fft


class Fun:
    def __init__(self, start=0, end=0, step=0):
        self.initX(start, end, step)
        self.initY()

    def initX(self, start, end, step):
        self.start = start
        self.end = end
        self.step = step
        self.length = int(np.ceil(np.abs(end - start) / step))
        self.X = np.arange(start, end, step)

    def initY(self):
        self.Y = np.zeros(self.length)

    def spectr(self, func, dt):
        re, im = 0.0, 0.0
        for k in range(func.size):
            re += func[k] * np.cos(2 * np.pi * self.X * k / func.size)
            im += func[k] * np.sin(2 * np.pi * self.X * k / func.size)
        Re = (1 / func.size) * re
        Im = (1 / func.size) * im
        self.Y = np.sqrt(Re ** 2 + Im ** 2)
        df = 1 / (self.end * dt)
        self.initX(self.start, self.end * df, df)
        self.initX(self.start, int(self.end / 2), self.step)
        self.Y = np.array(list(self.Y)[:self.length])

def signal_am(amp=10.0, km=0.25, fc=10.0, fs=2.0, period=100):
    tt = 2.0 * np.pi * np.linspace(0, 1, period)
    return amp * (1 + km * np.cos(fs * tt)) * np.cos(fc * tt)


if __name__ == '__main__':
    N = 1000

    # Create AM-signal
    fs = 10  # Modulation frequency
    fc = [25, 100, 150]  # Carrier frequency

    sig = [signal_am(amp=20.0, km=0.12, fc=i, fs=fs, period=N) for i in fc]

    # Calculate FFT
    sft = np.abs(fft(sig, axis=1)) / N / 0.5

    plt.figure(figsize=(20, 10), dpi=80)
    for i, freq in enumerate(fc):
        plt.subplot(len(fc), 2, 2 * i + 1)
        if i == 0:
            plt.title(' ')
        plt.plot(sig[i])
        plt.xlim([0, N - 1])
        plt.grid(False)

        plt.subplot(len(fc), 2, 2 * i + 2)
        if i == 0:
            plt.title('')
        plt.plot(sft[i], color='r')
        plt.xlim([0, N // 2 - 1])
        plt.grid(False)
    plt.tight_layout()
    plt.show()

    n = 0  # signal number

    sig_ns = numpy.convolve(sig[n], numpy.random.normal(size=sig[0].size), 'same')
    fn = Fun(0, sig_ns.size, 1)
    fn.spectr(sig_ns, 1 / sig_ns.size)

    plt.figure(figsize=(15, 5), dpi=80)
    plt.title(" ")
    plt.plot(sig_ns, '-', linewidth=1.0)
    plt.grid(False)
    plt.show()

    plt.figure(figsize=(15, 5), dpi=80)
    plt.title('  ')
    plt.plot(fn.Y, '-', linewidth=1.0)
    plt.grid(False)
    plt.show()

    sig_ns = np.array(sig[n])
    for i in range(300, 400):
        sig_ns[i] = 2 * np.random.rand() - 1

    fn = Fun(0, sig_ns.size, 1)
    fn.spectr(sig_ns, 1 / sig_ns.size)

    plt.title("")
    plt.plot(sig_ns, '-', linewidth=2.0)
    plt.grid(True)
    plt.show()

    plt.title('')
    plt.plot(fn.Y, '-', linewidth=2.0)
    plt.grid(True)
    plt.show()

```
